# Visual Odometry
This script is for finding and ploting the estimated path from the rotation and translation matricies computed in OpenCV C++ code

```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import animation as anm
from matplotlib import image as img
from numpy.linalg import inv
#%matplotlib inline
```

## Ground Truth Data

```python
# Load truth data from file
vo_practice_truth = np.genfromtxt('../data/vo_practice_sequence_truth.txt', delimiter=' ')

# Load truth data into list of matricies
motion = []
for row in vo_practice_truth:
    M = np.empty([4, 4])
    M[0, 0] = row[0];
    M[0, 1] = row[1];
    M[0, 2] = row[2];
    M[0, 3] = row[3];
    M[1, 0] = row[4];
    M[1, 1] = row[5];
    M[1, 2] = row[6];
    M[1, 3] = row[7];
    M[2, 0] = row[8];
    M[2, 1] = row[9];
    M[2, 2] = row[10];
    M[2, 3] = row[11];
    M[3, 0] = 0;
    M[3, 1] = 0;
    M[3, 2] = 0;
    M[3, 3] = 1;
    motion.append(M)
```

```python
# Start from point 0,0
X = np.array([0, 0, 0, 1])
X.shape = (4,1)

# Create lists for x and y coordinates
x_truth = []
y_truth = []
z_truth = []

# Iterate over motion
for M in motion:
    P = np.matmul(M, X)
    x_truth.append(P[0,0])
    y_truth.append(P[1,0])
    z_truth.append(P[2,0])
```

## Visual Odometry Data

```python
# Load odometry data from file
vo_practice_data = np.genfromtxt('../output/vo_practice_sequence_data.txt', delimiter=' ')

# Load truth data into list of matricies
motion = []
for row in vo_practice_data:
    M = np.empty([4, 4])
    M[0, 0] = row[0];
    M[0, 1] = row[1];
    M[0, 2] = row[2];
    M[0, 3] = row[3];
    M[1, 0] = row[4];
    M[1, 1] = row[5];
    M[1, 2] = row[6];
    M[1, 3] = row[7];
    M[2, 0] = row[8];
    M[2, 1] = row[9];
    M[2, 2] = row[10];
    M[2, 3] = row[11];
    M[3, 0] = 0;
    M[3, 1] = 0;
    M[3, 2] = 0;
    M[3, 3] = 1;
    motion.append(M)
```

```python
# Start from point 0,0
X = np.array([0, 0, 0, 1])
X.shape = (4,1)

# Create lists for x and y coordinates
x_odometry = []
y_odometry = []
z_odometry = []

# Iterate over motion
for M in motion:
    X = np.matmul(M, X)
    #X = np.matmul(inv(M), X)
    x_odometry.append(X[2,0])
    y_odometry.append(X[1,0])
    z_odometry.append(X[0,0])
```

## Plot Results

```python
# Plot data
fig, (ax) = plt.subplots(1, 1, figsize=(9,6))
ax.plot(x_truth, z_truth, 'ro', label='Ground Truth')
ax.plot(x_odometry, z_odometry, 'bo', label='Odometry Data')

# Label plot
ax.set_title('Visual Odometry Track')
ax.set_xlabel('x')
ax.set_ylabel('z')
ax.legend()

plt.show()
```

```python
def image_sequence_resolve_all(filepath):
    import os

    basedir, filename = os.path.split(filepath)
    filename_noext, ext = os.path.splitext(filename)

    from string import digits
    if isinstance(filepath, bytes):
        digits = digits.encode()
    filename_nodigits = filename_noext.rstrip(digits)

    if len(filename_nodigits) == len(filename_noext):
        # input isn't from a sequence
        return []

    files = os.listdir(basedir)
    return [
        os.path.join(basedir, f)
        for f in files
        if f.startswith(filename_nodigits) and
           f.endswith(ext) and
           f[len(filename_nodigits):-len(ext) if ext else -1].isdigit()]

image_list = image_sequence_resolve_all("../output/vo_practice_sequence/000001.png")
```
