# Ball Tracker Post Processing

The following is the computation of the trajectory from processed points for the baseball catcher project

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# Create data of ball points
ball = np.array( [
        [24.2554, -27.0417, 408.737],
        [23.9705, -27.1068, 397.578],
        [23.617, -27.0489, 386.01],
        [23.2192, -26.9779, 373.542],
        [22.9437, -26.762, 362.565],
        [22.639, -26.5592, 351.931],
        [22.2792, -26.2489, 340.085],
        [21.9925, -25.9843, 330.005],
        [21.6495, -25.5479, 318.04],
        [21.3661, -25.1002, 307.576],
        [21.093, -24.711, 297.363],
        [20.8239, -24.1101, 286.032],
        [20.5372, -23.4801, 275.1],
        [20.2846, -22.8109, 264.379],
        [20.0348, -22.0578, 253.559],
        [19.8039, -21.2955, 242.826],
        [19.5776, -20.494, 232.429],
        [19.3534, -19.5811, 221.929],
        [19.1238, -18.644, 211.281],
        [18.8843, -17.5919, 200.955],
        [18.687, -16.7063, 191.259],
        [18.4944, -15.6874, 180.725],
        [18.3378, -13.878, 167.453],
        [18.1393, -12.7147, 159.725],
        [17.9496, -11.6262, 151.844],
        [17.8143, -10.2729, 142.215],
        [17.7247, -8.84575, 132.534],
        [17.5633, -7.33512, 122.657],
        [17.3978, -5.7759, 112.448],
        [17.2093, -4.09655, 102.098],
        [17.2837, -2.41127, 93.19],
        [17.4271, -0.669638, 84.4581]])

# Create x, y, and z from data
x = ball[:,0] - 10
y = ball[:,1]
z = ball[:,2]
```

## Predict Trajectory in X

The trajectory prediction is done using least squares. Specifically, I use a quadtratic data fit. The problem is setup as follows. Our quadratic equation is

$az^{2} + bz + c$.

We create matricies $A$, $x$, and $b$ to match this equation.

$A =
\begin{bmatrix}
    z_{1}^{2} & z_{1} & 1 \\
    z_{2}^{2} & z_{2} & 1 \\
    \vdots    & \vdots & \vdots \\
    z_{N}^{2} & z_{N} & 1
\end{bmatrix}
,
x =
\begin{bmatrix}
    a \\
    b \\
    c
\end{bmatrix}
,
b =
\begin{bmatrix}
    x_{1} \\
    x_{2} \\
    \vdots \\
    x_{N}
\end{bmatrix}$

This forms the equation

$Ax = b$

This is an over-detirmined problem and thus the least-squares solution can be applied

$x = (A^{T}A)^{-1}A^{T}x$

Computation and results follow

```python
# Setup A and b
A = np.vstack([z**2, z, np.ones(len(z))]).T
b = x

# Compute least squares solution
x_hat = np.linalg.lstsq(A,b)[0]
print(x_hat)
z_rng = np.arange(0, 450, 0.1)
x_est = x_hat[0] * z_rng**2 + x_hat[1] * z_rng + x_hat[2] * np.ones(len(z_rng))
print('Estimated final x position')
print(x_est[0])
```

```python
# Plot x versus z data
fig, (ax) = plt.subplots(1, 1, figsize=(9,6))
ax.plot(z, x, label='Measured Trajectory')
ax.plot(z_rng, x_est, label='Estimated Trajectory')
ax.set_title("Ball trajectory in x")
ax.set_xlabel('z')
ax.set_ylabel('x')
ax.invert_xaxis()
ax.legend()
plt.show()

# Save figure
fig.savefig('../output/ball_trajectory_x.png', bbox_inches='tight')
```

## Trajectory Prediction in y

I approach this, once again, as a quadratic least squares problem. This is especially fitting since the acceloration, velocity, position model is qudratic. The matricies are setup as

$A =
\begin{bmatrix}
    z_{1}^{2} & z_{1} & 1 \\
    z_{2}^{2} & z_{2} & 1 \\
    \vdots    & \vdots & \vdots \\
    z_{N}^{2} & z_{N} & 1
\end{bmatrix}
,
x =
\begin{bmatrix}
    a \\
    b \\
    c
\end{bmatrix}
,
b =
\begin{bmatrix}
    y_{1} \\
    y_{2} \\
    \vdots \\
    y_{N}
\end{bmatrix}$

```python
# Setup b (A already setup)
b = y

# Compute least squares solution
x_hat = np.linalg.lstsq(A,b)[0]
print(x_hat)
z_rng = np.arange(0, 450, 0.1)
y_est = x_hat[0] * z_rng**2 + x_hat[1] * z_rng + x_hat[2] * np.ones(len(z_rng))
print('Estimated final y position')
print(y_est[0])
```

```python
# Plot y versus z
fig, (ax) = plt.subplots(1, 1, figsize=(9,6))
ax.plot(z, y, label='Measured Trajectory')
ax.plot(z_rng, y_est, label='Estimated Trajectory')
ax.set_title("Ball trajectory in y")
ax.set_xlabel('z')
ax.set_ylabel('y')
ax.invert_xaxis()
ax.invert_yaxis()
ax.legend()
plt.show()

# Save figure
fig.savefig('../output/ball_trajectory_y.png', bbox_inches='tight')
```
